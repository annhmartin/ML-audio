```python
import laion_clap

model = laion_clap.CLAP_Module(enable_fusion=False)
model.load_ckpt()   # downloads the pretrained checkpoint (~2 GB, one time)
print('CLAP loaded.')
```

```python
one = wavs[0]
emb = model.get_audio_embedding_from_filelist(x=[one], use_tensor=False)
print('Clip:', os.path.basename(one))
print('Embedding shape:', emb.shape)   # expect (1, 512)
print('First 5 values:', emb[0][:5])
assert emb.shape == (1, 512), 'Unexpected embedding shape'
print('\n✅ Backbone works.')
```

```python
import pandas as pd, numpy as np, faiss

cap = pd.read_csv(os.path.join(DRIVE_DIR, 'clotho_captions_development.csv'), dtype=str)
cap = cap.set_index('file_name')
cap_cols = [c for c in cap.columns if c.lower().startswith('caption')]

name_to_path = {os.path.basename(p): p for p in wavs}
subset = [n for n in cap.index if n in name_to_path][:30]
paths = [name_to_path[n] for n in subset]
print(f'Embedding {len(paths)} clips...')

X = model.get_audio_embedding_from_filelist(x=paths, use_tensor=False).astype('float32')
faiss.normalize_L2(X)
index = faiss.IndexFlatIP(X.shape[1])
index.add(X)

D, I = index.search(X[0:1], 3)
q = subset[0]; nn = subset[I[0][1]]
print('\nQUERY  :', q)
print('  caption:', cap.loc[q, cap_cols[0]])
print('\nNEAREST:', nn, f'(similarity {D[0][1]:.3f})')
print('  caption:', cap.loc[nn, cap_cols[0]])
print('\nDo the two captions describe sounds? If yes, retrieval is sane. But might not be accurate yet.')
```

```python
import numpy as np, json, os
CACHE_DIR  = os.path.join(DRIVE_DIR, 'cache')
emb_path   = os.path.join(CACHE_DIR, 'dev_embeddings.npy')
names_path = os.path.join(CACHE_DIR, 'dev_filenames.json')

if os.path.exists(emb_path):
    X_all = np.load(emb_path)
    all_names = json.load(open(names_path))
    print('Loaded cached embeddings instantly:', X_all.shape)
else:
    print('No cache yet — run the full embedding cell once.')
```

```python
import faiss
Xn = X_all.copy()
faiss.normalize_L2(Xn)
full_index = faiss.IndexFlatIP(Xn.shape[1])
full_index.add(Xn)
faiss.write_index(full_index, os.path.join(CACHE_DIR, 'dev.index'))
print('Index built and saved:', full_index.ntotal, 'vectors')
```

```python
import numpy as np, random
Xn = X_all.astype('float32').copy()
faiss.normalize_L2(Xn)

random.seed(0)
for qi in random.sample(range(len(all_names)), 5):
    D, I = full_index.search(Xn[qi:qi+1], 3)
    q = all_names[qi]
    print('='*70)
    print('QUERY :', q)
    print('   ->', cap.loc[q, cap_cols[0]])
    for rank in [1, 2]:                 # rank 0 is the clip itself, skip it
        nn = all_names[I[0][rank]]
        print(f'NN{rank} (sim {D[0][rank]:.3f}):', nn)
        print('   ->', cap.loc[nn, cap_cols[0]])
```
