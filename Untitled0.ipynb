```python
# @title STAGE 1: Create 19K training pairs/tokenizer/padding/max length/CLAP embedding

# build readable (filename, caption) pairs, with a NaN/empty guard
# since it is 5 captions per audio each caption is made into a pair with the corresponding audio
# So like 19K training pairs
  # 3,839 audio files * 5 captions = 19,195 training pairs
training_pairs = []
for audio_name in all_names:
    if audio_name in cap.index:
        for col in cap_cols:
            caption_text = cap.loc[audio_name, col]
            if isinstance(caption_text, str) and caption_text.strip():
                training_pairs.append({'audio_filename': audio_name,
                                       'caption': caption_text.strip()})
training_df = pd.DataFrame(training_pairs)
print(f'Generated {len(training_df)} training pairs.')
display(training_df.head())

# tokenizer + map each row to its (normalized) embedding

# Loading GPT2's tokenizer, converts text into integer sequence that can be processed
#since there isn't a padding token, the end-of-sequence token is used as the padding token
tok = GPT2Tokenizer.from_pretrained('gpt2')
tok.pad_token = tok.eos_token
#look up dictionary with filename to the row index
name_to_row = {n: i for i, n in enumerate(all_names)}   # filename -> row in X_all
MAX_LEN = 30 #max length with padding or truncation

#fetching, each row returns:
  # 1)emb --> CLAP embedding converted to PyTorch tensor; input for the model
  # 2)enc.input_ids --> caption converted to sequence of token integers to max length; target to produce
  # 3)end.attention_mask --> denotes real tokens/padding with 1s/0s; hence ignore padding during training
class CapDataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        r = self.df.iloc[idx]
        emb = torch.tensor(X_all[name_to_row[r['audio_filename']]], dtype=torch.float32)
        enc = tok(r['caption'] + tok.eos_token, truncation=True, max_length=MAX_LEN,
                  padding='max_length', return_tensors='pt')
        return emb, enc.input_ids[0], enc.attention_mask[0]

#Dataloader is auto batching (64) and shuffling (randomized each epoch)
ds = CapDataset(training_df)
dl = DataLoader(ds, batch_size=64, shuffle=True)
emb, ids, m = next(iter(dl))
print('Batch -> embeddings:', emb.shape, '| tokens:', ids.shape)
```

```python
#  @title STAGE 2: Model Architecture

# defines two PyTorch nn.Module classes
import torch, torch.nn as nn

PREFIX_LEN, GPT2_DIM = 10, 768

#MappingNetwork - is a two layer Multi-layer Perceptron, using Tahn non-linearity
# Translate embedding into 10 tokens of 768 numbers each for GPT2 so it can understand it
  # Layer 1 is going from 512 to 3840 numbers
  # Tanh changes all values to be between -1 and 1 with non linearity to make comples patterns and not linear translation
  # Layer 2 going from 3840 to 7680
# Then finally the .view(-1, 10, 768) reshapes the 7680 into 10 tokens of 768 dimensions each which is the prefix
class MappingNetwork(nn.Module):
    def __init__(self, clap_dim=512, prefix_len=PREFIX_LEN, dim=GPT2_DIM):
        super().__init__()
        self.prefix_len, self.dim = prefix_len, dim
        self.net = nn.Sequential(
            nn.Linear(clap_dim, dim * prefix_len // 2), # (512, 768*10/2) => (512, 3840)
            nn.Tanh(),
            nn.Linear(dim * prefix_len // 2, dim * prefix_len), # (768*10/2, 768*10) => (3840, 7680)
        )
    def forward(self, x):                          # x: (B, 512)
        return self.net(x).view(-1, self.prefix_len, self.dim)   # (B, P, 768)

# System A model defined (AudioCaptioner)
# Loads the pretrained GPT2 and freezes it so that it's langague abilities are preserved
# Only MappingNetwoek parameters are trainable - prefix tuning (fine tuning of only the small adapter network)
class AudioCaptioner(nn.Module):
    def __init__(self):
        super().__init__()
        self.gpt = GPT2LMHeadModel.from_pretrained('gpt2')
        for p in self.gpt.parameters():                       # freeze GPT-2
            p.requires_grad = False
        self.map = MappingNetwork()

    # self.map(emb) converts the audio embedding into 10 prefix tokens
    # wte(input_ids) converts caption text into 30 token embedding (Word Token Embedding)
    # torch.cat sticks the into a 40 token sequence (10 audio tokens and 30 text tokens)
    #the model then reads the sequence l>r and uses the ausio prefix to predict the caption tokens that will follow
    def forward(self, emb, input_ids, attn_mask):
        prefix    = self.map(emb)                              # (B, P, 768)
        tok_embed = self.gpt.transformer.wte(input_ids)        # (B, L, 768)
        inputs    = torch.cat([prefix, tok_embed], dim=1)      # (B, P+L, 768)

        B, P = prefix.shape[0], prefix.shape[1]
        prefix_mask = torch.ones(B, P, device=emb.device, dtype=attn_mask.dtype)
        mask = torch.cat([prefix_mask, attn_mask], dim=1)
        #Model penalization defined, -100
        #Peanlization only on actual caption words, not padding
        labels = input_ids.clone()
        labels[attn_mask == 0] = -100                          # ignore padding
        prefix_labels = torch.full((B, P), -100, device=emb.device, dtype=labels.dtype)
        labels = torch.cat([prefix_labels, labels], dim=1)     # ignore prefix too
        out = self.gpt(inputs_embeds=inputs, attention_mask=mask, labels=labels)
        return out.loss

device = 'cuda' if torch.cuda.is_available() else 'cpu'
captioner = AudioCaptioner().to(device)

emb, ids, m = next(iter(dl))
loss = captioner(emb.to(device), ids.to(device), m.to(device))
n_train = sum(p.numel() for p in captioner.parameters() if p.requires_grad)
print(f'Trainable params: {n_train:,}  (GPT-2 is frozen)')
print('Sanity loss on one batch:', loss.item())

#showing 31M trainable params on MappingNetwork
#Loss of 9.029 is pretty much random right now since training has not yet happened
```

```python
# @title Caption_For - inference function for System A

# Usng beam search decoding for System A
@torch.no_grad() # disables gradient computation throughout; speeding up inference by halving the memory usage
def caption_for(emb_vec, max_len=30):
    captioner.eval()
    v = np.asarray(emb_vec, dtype='float32') #raw CLAP embedding to float32
    v = v / (np.linalg.norm(v) + 1e-9) #normalize, prevent divide-by-zero
    emb = torch.tensor(v, device=device).unsqueeze(0) #(512,) -> (1,512)
    prefix = captioner.map(emb) #translate audio into 10 prefix tokens GPT-2 can read
    # beam search instead of greedy argmax: explores 5 candidate sentences and keeps the best
    # no_repeat_ngram_size=2 blocks repeated word pairs ("chirp and chirp")
    # reduces the risk of missing hidden highh probability word sequences
    attn_mask = torch.ones(prefix.shape[:2], dtype=torch.long, device=device)
    out = captioner.gpt.generate(
        inputs_embeds=prefix,
        attention_mask=attn_mask,   #to deal with gpt'sinabilitxy to infer mask; w/o then all positions are essentailly equal, including padding
        max_new_tokens=max_len,
        num_beams=5,                #5 candidate sequence at once, selects the highest cumulative log-probability sequence at completion
        no_repeat_ngram_size=2,     #blocks the bigrams from showing up two times; eliminating the whole repetition issue (ie 'a train passing and a train passing#)
        early_stopping=True,
        pad_token_id=tok.eos_token_id)
    # token ids into human-readable string, drop special tokens, strip whitespace
    return tok.decode(out[0], skip_special_tokens=True).strip()

#Quick review of above with 5 random clips
for n in random.sample(all_names, 5):
    print('CLIP   :', n)
    print('  MODEL:', caption_for(X_all[name_to_row[n]]))
    print('  TRUE :', cap.loc[n, cap_cols[0]], '\n')
```

```python
#  @title B1 Hint builder; In-context Learning

#Creating hints for every clip with accoustically similar 3 clips
# batch-searching the FAISS index
# this in-context learning
  # pattern recognition procedure than actual learning procedure
  # allows the model to look at similar examples it was exposed to during training
  # previous research shows that there can be some instability with this technique
K = 3
hint_for = {}
Xn_all = X_all.astype('float32').copy(); faiss.normalize_L2(Xn_all)
Dn, In = full_index.search(Xn_all, K + 1)          # retrieve neighbors for all clips at once
for i, name in enumerate(all_names):
    neigh = [all_names[j] for j in In[i] if all_names[j] != name][:K] #Filters out any self-matches
    # the 'Caption' at the end is a prompt boundary for gpt2 so it will create a caption and not just cntinue with a description
    hint_for[name] = 'Similar sounds: ' + ' '.join(cap.loc[nm, cap_cols[0]] for nm in neigh) + ' Caption:'
print('Example hint:\n', hint_for[all_names[0]])
```

```python
# @title B2 RagDataset

#bundling each training with three things:
  #1) audio embedding
  #2) true caption (caption token IDs, caption attention mask)
  #3) hint string (hint token IDs, hint attention mask)
HINT_LEN, CAP_LEN = 64, 30 # these are tokenized separately for independent masking
class RagDataset(Dataset):
    def __init__(self, df): self.df = df.reset_index(drop=True)
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        r = self.df.iloc[idx]
        emb = torch.tensor(X_all[name_to_row[r['audio_filename']]], dtype=torch.float32)
        h = tok(hint_for[r['audio_filename']], truncation=True, max_length=HINT_LEN,
                padding='max_length', return_tensors='pt')
        c = tok(r['caption'] + tok.eos_token, truncation=True, max_length=CAP_LEN,
                padding='max_length', return_tensors='pt')
        return emb, h.input_ids[0], h.attention_mask[0], c.input_ids[0], c.attention_mask[0]
rag_dl = DataLoader(RagDataset(training_df), batch_size=64, shuffle=True)
```

```python
# @title B3 RagCaptioner - defining System B model

# New translator being trained to use the hints, captions, and audio
class RagCaptioner(nn.Module):
    def __init__(self):
        super().__init__()
        self.gpt = captioner.gpt     # same frozen GPT-2 as System A
        self.map = MappingNetwork()  # NEW translator, trained w/hints; it is independently initialized MappingNetwork
    def forward(self, emb, hids, hmask, cids, cmask):
        prefix = self.map(emb)
        he = self.gpt.transformer.wte(hids); ce = self.gpt.transformer.wte(cids)
        inp = torch.cat([prefix, he, ce], dim=1)
        B, P = prefix.shape[0], prefix.shape[1]
        pmask = torch.ones(B, P, device=emb.device, dtype=hmask.dtype)
        mask = torch.cat([pmask, hmask, cmask], dim=1)
        labels = cids.clone(); labels[cmask == 0] = -100
        ignore = torch.full((B, P + hids.shape[1]), -100, device=emb.device, dtype=labels.dtype) # loss masking on prefix and hint region
        labels = torch.cat([ignore, labels], dim=1)            # score only the 30 caption positions; keeps it from predicting hint tex it already has, like covering the answer key when studying
        return self.gpt(inputs_embeds=inp, attention_mask=mask, labels=labels).loss
```

```python
# @title caption functions for held-out clips
# (System B retrieves from DEV)
# Retrieval baseline is created here

#Four inference functions that are used for evaluating
  # 1) hint_from_dev
  # 2) sysA_emb
  # 3) sysB_emb
  # 4) retrieval_emb

# From the training FAISS index gives the formatted hint string and the neighbors
  # this is for retrieval-echo implementation
def hint_from_dev(emb_vec, k=3):
    v = emb_vec.astype('float32').copy(); v = v/(np.linalg.norm(v)+1e-9)
def hint_from_dev(emb_vec, k=3):
    v = emb_vec.astype('float32').copy(); v = v/(np.linalg.norm(v)+1e-9)
    D,I = full_index.search(v.reshape(1,-1), k)
    neigh = [all_names[j] for j in I[0]][:k]
    return 'Similar sounds: ' + ' '.join(cap.loc[nm, cap_cols[0]] for nm in neigh) + ' Caption:', neigh

#separates the embedding input from the function implementation
def sysA_emb(emb_vec):  # audio only essentially
    return caption_for(emb_vec)

#gets all of System B, like every part of it; inference function defined as well for B
@torch.no_grad()
def sysB_emb(emb_vec, max_len=30):
    rag_model.eval()
    v = emb_vec.astype('float32')
    v = v / (np.linalg.norm(v) + 1e-9)
    emb = torch.tensor(v, device=device).unsqueeze(0)

    # Generate hint
    hint, _ = hint_from_dev(emb_vec)
    hint_enc = tok(
        hint,
        truncation=True,
        max_length=HINT_LEN,
        padding='max_length',
        return_tensors='pt'
    )
    hids = hint_enc.input_ids.to(device)
    hint_mask = hint_enc.attention_mask.to(device)

    # Generate prefix from audio embedding
    prefix = rag_model.map(emb)

    # Get embeddings for hint and prefix
    he = rag_model.gpt.transformer.wte(hids)
    inp = torch.cat([prefix, he], dim=1)

    # Attention mask: 1 for real tokens, 0 for padding
    prefix_mask = torch.ones(prefix.shape[:2], dtype=torch.long, device=device)
    attn_mask = torch.cat([prefix_mask, hint_mask], dim=1)

    # Generate caption
    out = rag_model.gpt.generate(
        inputs_embeds=inp,
        attention_mask=attn_mask,
        max_new_tokens=max_len,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True,
        pad_token_id=tok.eos_token_id
    )

    return tok.decode(out[0], skip_special_tokens=True).strip()

# the baseline that just returns nesarest clip captions w/o generation
def retrieval_emb(emb_vec):
    _, neigh = hint_from_dev(emb_vec, k=1)
    return cap.loc[neigh[0], cap_cols[0]]

print(hint_from_dev)
print(sysA_emb)
print(sysB_emb)
print(retrieval_emb)
```

```python
# @title STAGE 4: metrics on HELD-OUT evaluation clips - 1045

#evaluating all three systems (Retrieval, A, B) on evaluation clips
  # using pycocoevalcap since it is the standard evaluation library for audio captioning
    # from the COCO dataset benchmark
  # each system generates 1 caption/clip and then scored against the 5 human references
def evaluate_system(label, caption_fn, names_list):
    refs, hyps = {}, {}
    print(f'Generating captions for {label} on {len(names_list)} clips...')
    for i, n in enumerate(names_list):
        emb = X_eval[eval_name_to_row[n]]
        hyps[i] = [caption_fn(emb)]
        refs[i] = [cap_eval.loc[n,c] for c in eval_cap_cols
                   if isinstance(cap_eval.loc[n,c], str) and cap_eval.loc[n,c].strip()]
    out={}
    for scorer,nm in [(Bleu(4),['BLEU-1','BLEU-2','BLEU-3','BLEU-4']),
                      (Meteor(),['METEOR']),(Cider(),['CIDEr'])]:
        sc,_ = scorer.compute_score(refs,hyps)
        if isinstance(sc,list):
            for a,b in zip(nm,sc): out[a]=round(b,4)
        else: out[nm[0]]=round(sc,4)
    return out

random.seed(42)
test_names = random.sample(eval_names, min(1045, len(eval_names)))

results = {}
results['Retrieval (NN)'] = evaluate_system('Retrieval', retrieval_emb, test_names)
results['System A']       = evaluate_system('System A',  sysA_emb,      test_names)
results['System B (RAG)'] = evaluate_system('System B',  sysB_emb,      test_names)

df_results = pd.DataFrame(results).T[['BLEU-1','BLEU-2','BLEU-3','BLEU-4','METEOR','CIDEr']]
display(df_results)
```

```python
# @title BERT Scoring

#BERTScores using distilbert-base-uncased
#for each of the pairs the contextual token embedding are computed
  #the pairwise cosine similarities are used to compute
    #pricision
    #recall
    #F1

#added this scoring to the above scoring since BLEU is simply n-gram matching
  #and BERT is looking at semantic equivalence
  #ie "a stream flows quickly" vs "water runs gently in a brook" would score well with BERT

def evaluate_bertscore(label, caption_fn, names_list):
    print(f'Computing BERTScore for {label} on {len(names_list)} clips...')
    hyps = []
    refs = []  # BERTScore takes one reference per hypothesis
                # we use caption_1 as the reference
    for n in names_list:
        emb = X_eval[eval_name_to_row[n]]
        hyps.append(caption_fn(emb))
        refs.append(cap_eval.loc[n, eval_cap_cols[0]])

    P, R, F1 = bert_score(
        hyps, refs,
        lang='en',
        model_type='distilbert-base-uncased',  # lighter model, faster
        verbose=False
    )
    return {
        'BERTScore-P' : round(P.mean().item(), 4),
        'BERTScore-R' : round(R.mean().item(), 4),
        'BERTScore-F1': round(F1.mean().item(), 4),
    }

# run on same test_names as your existing metrics
bert_results = {}
bert_results['Retrieval (NN)'] = evaluate_bertscore('Retrieval', retrieval_emb, test_names)
bert_results['System A']       = evaluate_bertscore('System A',  sysA_emb,      test_names)
bert_results['System B (RAG)'] = evaluate_bertscore('System B',  sysB_emb,      test_names)

df_bert = pd.DataFrame(bert_results).T
print('\nBERTScore results')
display(df_bert.style.format('{:.4f}')
        .set_caption('BERTScore on held-out Clotho evaluation split'))
```

```python
# @title STAGE 5 Error analysis on HELD-OUT clips

#generating captins for 60 evaluation clips
# computes the per-clip overlap scores
#takes the max across all 5 references

#showing where RAG helped and where it hurt
random.seed(7)
analysis_names = random.sample(eval_names, 60)

def overlap(pred, ref):
    p, r = set(pred.lower().split()), set(ref.lower().split())
    return len(p & r) / max(len(r), 1)

rows = []
for n in analysis_names:
    emb  = X_eval[eval_name_to_row[n]]
    refs = [cap_eval.loc[n,c] for c in eval_cap_cols
            if isinstance(cap_eval.loc[n,c], str) and cap_eval.loc[n,c].strip()]
    a = sysA_emb(emb); b = sysB_emb(emb)
    hint, neigh = hint_from_dev(emb)
    nn_cap = cap.loc[neigh[0], cap_cols[0]]
    oa = max(overlap(a,r) for r in refs)            # best match vs any reference
    ob = max(overlap(b,r) for r in refs)
    echo = overlap(b, nn_cap)                       # how much B copied its nearest hint
    rows.append({'file':n,'true':refs[0],'A':a,'B':b,
                 'oa':round(oa,3),'ob':round(ob,3),
                 'echo':round(echo,3),'b_gain':round(ob-oa,3)})
df_err = pd.DataFrame(rows)

print('\nMean overlap   A: %.3f    B: %.3f' % (df_err.oa.mean(), df_err.ob.mean()))
print('B beats A on   %d/%d clips' % ((df_err.b_gain>0).sum(), len(df_err)))
#mean word overlap between System B generated captions and NN captions
print('Mean retrieval-echo (B vs nearest hint): %.3f' % df_err.echo.mean())

print('\n\n Where retrieval HELPED most (B >> A)')
for _,r in df_err.nlargest(3,'b_gain').iterrows():
    print(f'\nTRUE: {r.true}\n  A : {r.A}\n  B : {r.B}   (+{r.b_gain})')

print('\n\n Where retrieval HURT most (A >> B)')
for _,r in df_err.nsmallest(3,'b_gain').iterrows():
    print(f'\nTRUE: {r.true}\n  A : {r.A}\n  B : {r.B}   ({r.b_gain})')
```

```python
# @title Results Charts
warnings.filterwarnings('ignore')

tracker_path = os.path.join(CACHE_DIR, 'run_tracker.csv')
tracker = pd.read_csv(tracker_path).sort_values('timestamp').reset_index(drop=True)
tracker['run'] = range(len(tracker))

baselines = {
    'BLEU-1'      : {'clotho': 0.378, 'beam': 0.481,  'col_a': 'bleu1_a',        'col_b': 'bleu1_b',        'col_r': 'bleu1_ret',        'ylim': (0.2,  0.65), 'bert_ref': None},
    'BLEU-4'      : {'clotho': 0.017, 'beam': 0.115,  'col_a': 'bleu4_a',        'col_b': 'bleu4_b',        'col_r': None,               'ylim': (0.0,  0.20), 'bert_ref': None},
    'METEOR'      : {'clotho': 0.078, 'beam': 0.174,  'col_a': 'meteor_a',       'col_b': 'meteor_b',       'col_r': None,               'ylim': (0.05, 0.25), 'bert_ref': None},
    'CIDEr'       : {'clotho': 0.075, 'beam': 0.290,  'col_a': 'cider_a',        'col_b': 'cider_b',        'col_r': 'cider_ret',        'ylim': (0.0,  0.40), 'bert_ref': None},
    'BERTScore-F1': {'clotho': None,  'beam': None,   'col_a': 'bertscore_f1_a', 'col_b': 'bertscore_f1_b', 'col_r': 'bertscore_f1_ret', 'ylim': (0.75, 0.85), 'bert_ref': 0.80},
    'BERTScore-P' : {'clotho': None,  'beam': None,   'col_a': 'bertscore_p_a',  'col_b': 'bertscore_p_b',  'col_r': 'bertscore_p_ret',  'ylim': (0.75, 0.85), 'bert_ref': 0.80},
    'BERTScore-R' : {'clotho': None,  'beam': None,   'col_a': 'bertscore_r_a',  'col_b': 'bertscore_r_b',  'col_r': 'bertscore_r_ret',  'ylim': (0.75, 0.85), 'bert_ref': 0.80},
}

COL_A   = '#378ADD'
COL_B   = '#7F77DD'
COL_RET = '#D85A30'
COL_C   = '#888780'
COL_BM  = '#1D9E75'
COL_BR  = '#E0A020'   # BERTScore reference line color

def style_ax(ax, fig):
    fig.patch.set_facecolor('#1a1a1a')
    ax.set_facecolor('#1a1a1a')
    for spine in ax.spines.values():
        spine.set_edgecolor('#333333')

# CHART 1: line chart over runs
def plot_line(metric):
    cfg   = baselines[metric]
    col_a = cfg['col_a']
    col_b = cfg['col_b']

    fig, ax = plt.subplots(figsize=(10, 4.5))
    style_ax(ax, fig)

    runs   = tracker['run'].values
    vals_a = tracker[col_a].values if col_a in tracker.columns else np.full(len(runs), np.nan)
    vals_b = tracker[col_b].values if col_b in tracker.columns else np.full(len(runs), np.nan)

    ax.plot(runs, vals_a, color=COL_A, linewidth=2.5, marker='o', markersize=6, label='System A (direct)')
    ax.plot(runs, vals_b, color=COL_B, linewidth=2.5, marker='s', markersize=6, label='System B (RAG)')

    # retrieval line from tracker
    if cfg['col_r'] and cfg['col_r'] in tracker.columns:
        ret_vals = tracker[cfg['col_r']].dropna()
        if len(ret_vals) > 0:
            ret_val = ret_vals.iloc[-1]
            ax.axhline(ret_val, color=COL_RET, linewidth=1.5, linestyle='--',
                       label=f'Retrieval baseline ({ret_val:.3f})')

    # published baselines
    if cfg['clotho']:
        ax.axhline(cfg['clotho'], color=COL_C, linewidth=1.2, linestyle=':',
                   label=f'Clotho V2 baseline ({cfg["clotho"]:.3f})')
    if cfg['beam']:
        ax.axhline(cfg['beam'], color=COL_BM, linewidth=1.2, linestyle=':',
                   label=f'Beam search comparable ({cfg["beam"]:.3f})')
    if cfg['bert_ref']:
        ax.axhline(cfg['bert_ref'], color=COL_BR, linewidth=1.2, linestyle=':',
                   label=f'Good captioning threshold ({cfg["bert_ref"]:.2f})')

    # data labels
    for r, a, b in zip(runs, vals_a, vals_b):
        if not np.isnan(a):
            ax.annotate(f'{a:.3f}', (r, a), textcoords='offset points',
                       xytext=(0, 8), fontsize=8, color=COL_A, ha='center')
        if not np.isnan(b):
            ax.annotate(f'{b:.3f}', (r, b), textcoords='offset points',
                       xytext=(0, -13), fontsize=8, color=COL_B, ha='center')

    run_labels = [f"Run {r}\n{tracker.loc[tracker['run']==r,'label'].values[0]}"
                  for r in runs]
    ax.set_xticks(runs)
    ax.set_xticklabels(run_labels, fontsize=8, color='#cccccc')
    ax.set_ylim(cfg['ylim'])
    ax.set_title(f'{metric} — System A & B across all runs', color='white', fontsize=13, pad=10)
    ax.set_ylabel(metric, color='#aaaaaa', fontsize=10)
    ax.set_xlabel('Run', color='#aaaaaa', fontsize=10)
    ax.tick_params(axis='y', colors='#aaaaaa')
    ax.grid(axis='y', color=(1, 1, 1, 0.07), linewidth=0.7)
    ax.legend(loc='lower right', fontsize=8, framealpha=0.3,
              labelcolor='white', facecolor='#2a2a2a', edgecolor='#444444')
    plt.tight_layout()
    plt.show()

# CHART 2: horizontal bar chart — most recent run
def plot_bars(metric):
    cfg  = baselines[metric]
    last = tracker.iloc[-1]

    val_a   = last[cfg['col_a']] if cfg['col_a'] in last.index else np.nan
    val_b   = last[cfg['col_b']] if cfg['col_b'] in last.index else np.nan
    val_ret = last[cfg['col_r']] if cfg['col_r'] and cfg['col_r'] in last.index else np.nan

    systems = ['System A\n(direct)', 'System B\n(RAG)', 'Retrieval\nbaseline']
    values  = [val_a, val_b, val_ret]
    colors  = [COL_A, COL_B, COL_RET]

    pairs = [(s, v, c) for s, v, c in zip(systems, values, colors) if not np.isnan(v)]
    systems, values, colors = zip(*pairs)

    fig, ax = plt.subplots(figsize=(9, 3.5))
    style_ax(ax, fig)

    bars = ax.barh(systems, values, color=colors, height=0.5, zorder=3)

    for bar, val in zip(bars, values):
        ax.text(val + max(values) * 0.01, bar.get_y() + bar.get_height() / 2,
                f'{val:.4f}', va='center', fontsize=9, color='white')

    xmax = max(values) * 1.2

    # retrieval as dotted vertical line if it's in the bar set,
    # also add it as a reference line regardless
    if not np.isnan(val_ret):
        ax.axvline(val_ret, color=COL_RET, linewidth=1.5, linestyle='--',
                   label=f'Retrieval baseline ({val_ret:.3f})', zorder=2)

    if cfg['clotho']:
        ax.axvline(cfg['clotho'], color=COL_C, linewidth=1.5, linestyle=':',
                   label=f'Clotho V2 baseline ({cfg["clotho"]:.3f})', zorder=2)
        xmax = max(xmax, cfg['clotho'] * 1.1)
    if cfg['beam']:
        ax.axvline(cfg['beam'], color=COL_BM, linewidth=1.5, linestyle=':',
                   label=f'Beam search comparable ({cfg["beam"]:.3f})', zorder=2)
        xmax = max(xmax, cfg['beam'] * 1.1)
    if cfg['bert_ref']:
        ax.axvline(cfg['bert_ref'], color=COL_BR, linewidth=1.5, linestyle=':',
                   label=f'Good captioning threshold ({cfg["bert_ref"]:.2f})', zorder=2)
        xmax = max(xmax, cfg['bert_ref'] * 1.05)

    ax.set_xlim(0, xmax)
    ax.set_title(
        f'{metric} — most recent run vs published baselines\n'
        f'(Run: {last["label"]} · {last["timestamp"]})',
        color='white', fontsize=12, pad=8)
    ax.set_xlabel(metric, color='#aaaaaa', fontsize=10)
    ax.tick_params(axis='x', colors='#aaaaaa')
    ax.tick_params(axis='y', colors='#cccccc', labelsize=9)
    ax.grid(axis='x', color=(1, 1, 1, 0.07), linewidth=0.7, zorder=0)
    ax.legend(loc='lower right', fontsize=8, framealpha=0.3,
              labelcolor='white', facecolor='#2a2a2a', edgecolor='#444444')
    plt.tight_layout()
    plt.show()

# interactive widgets
metric_opts = list(baselines.keys())

print('Chart 1 — Line chart: System A & B over all runs')
interact(plot_line, metric=Dropdown(options=metric_opts, value='CIDEr', description='Metric:'))

print('\nChart 2 — Bar chart: most recent run vs published baselines')
interact(plot_bars, metric=Dropdown(options=metric_opts, value='CIDEr', description='Metric:'))
```
