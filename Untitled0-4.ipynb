```python
# STAGE 1
import torch
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import GPT2Tokenizer

# build readable (filename, caption) pairs, with a NaN/empty guard
# since it is 5 captions per audio each caption is made into a pair with the corresponding audio
# So like 19K training pairs
training_pairs = []
for audio_name in all_names:
    if audio_name in cap.index:
        for col in cap_cols:
            caption_text = cap.loc[audio_name, col]
            if isinstance(caption_text, str) and caption_text.strip():
                training_pairs.append({'audio_filename': audio_name,
                                       'caption': caption_text.strip()})
training_df = pd.DataFrame(training_pairs)
print(f'Generated {len(training_df)} training pairs.')
display(training_df.head())

# tokenizer + map each row to its (normalized) embedding

# Loading GPT2's tokenizer, converts text into integer sequence that can be processed
#since there isn't a padding token, the end-of-sequence token is used as the padding token
tok = GPT2Tokenizer.from_pretrained('gpt2')
tok.pad_token = tok.eos_token
#look up dictionary with filename to the row index
name_to_row = {n: i for i, n in enumerate(all_names)}   # filename -> row in X_all
MAX_LEN = 30 #max length with padding or truncation

#fetching, each row returns:
  # 1)emb --> CLAP embedding converted to PyTorch tensor; input for the model
  # 2)enc.input_ids --> caption converted to sequence of token integers to max length; target to produce
  # 3)end.attention_mask --> denotes real tokens/padding with 1s/0s; hence ignore padding during training
class CapDataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        r = self.df.iloc[idx]
        emb = torch.tensor(X_all[name_to_row[r['audio_filename']]], dtype=torch.float32)
        enc = tok(r['caption'] + tok.eos_token, truncation=True, max_length=MAX_LEN,
                  padding='max_length', return_tensors='pt')
        return emb, enc.input_ids[0], enc.attention_mask[0]

#Dataloader is auto batching (64) and shuffling (randomized each epoch)
ds = CapDataset(training_df)
dl = DataLoader(ds, batch_size=64, shuffle=True)
emb, ids, m = next(iter(dl))
print('Batch -> embeddings:', emb.shape, '| tokens:', ids.shape)
```

```python
# STAGE 2: the model
import torch, torch.nn as nn
from transformers import GPT2LMHeadModel

PREFIX_LEN, GPT2_DIM = 10, 768

# Translate embedding into 10 tokens of 768 numbers each for GPT2 so it can understand it
  # Layer 1 is going from 512 to 3840 numbers
  # Tanh changes all values to be between -1 and 1 with non linearity to make comples patterns and not linear translation
  # Layer 2 going from 3840 to 7680
# Then finally the .view(-1, 10, 768) reshapes the 7680 into 10 tokens of 768 dimensions each which is the prefix
class MappingNetwork(nn.Module):
    def __init__(self, clap_dim=512, prefix_len=PREFIX_LEN, dim=GPT2_DIM):
        super().__init__()
        self.prefix_len, self.dim = prefix_len, dim
        self.net = nn.Sequential(
            nn.Linear(clap_dim, dim * prefix_len // 2), # (512, 768*10/2) => (512, 3840)
            nn.Tanh(),
            nn.Linear(dim * prefix_len // 2, dim * prefix_len), # (768*10/2, 768*10) => (3840, 7680)
        )
    def forward(self, x):                          # x: (B, 512)
        return self.net(x).view(-1, self.prefix_len, self.dim)   # (B, P, 768)

# Loads the pretrained GPT2 and freezes it so that it's langague abilities are preserved
# Only MappingNetwoek parameters are trainable
class AudioCaptioner(nn.Module):
    def __init__(self):
        super().__init__()
        self.gpt = GPT2LMHeadModel.from_pretrained('gpt2')
        for p in self.gpt.parameters():            # freeze GPT-2
            p.requires_grad = False
        self.map = MappingNetwork()

    # self.map(emb) converts the audio embedding into 10 prefix tokens
    # wte(input_ids) converts caption text into 30 token embedding (Word Token Embedding)
    # torch.cat sticks the into a 40 token sequence (10 audio tokens and 30 text tokens)
    #the model then reads the sequence l>r and uses the ausio prefix to predict the caption tokens that will follow
    def forward(self, emb, input_ids, attn_mask):
        prefix    = self.map(emb)                              # (B, P, 768)
        tok_embed = self.gpt.transformer.wte(input_ids)        # (B, L, 768)
        inputs    = torch.cat([prefix, tok_embed], dim=1)      # (B, P+L, 768)

        B, P = prefix.shape[0], prefix.shape[1]
        prefix_mask = torch.ones(B, P, device=emb.device, dtype=attn_mask.dtype)
        mask = torch.cat([prefix_mask, attn_mask], dim=1)
        #Model penalization defined, -100
        #Peanlization only on actual caption words, not padding
        labels = input_ids.clone()
        labels[attn_mask == 0] = -100                          # ignore padding
        prefix_labels = torch.full((B, P), -100, device=emb.device, dtype=labels.dtype)
        labels = torch.cat([prefix_labels, labels], dim=1)     # ignore prefix too
        out = self.gpt(inputs_embeds=inputs, attention_mask=mask, labels=labels)
        return out.loss

device = 'cuda' if torch.cuda.is_available() else 'cpu'
captioner = AudioCaptioner().to(device)

emb, ids, m = next(iter(dl))
loss = captioner(emb.to(device), ids.to(device), m.to(device))
n_train = sum(p.numel() for p in captioner.parameters() if p.requires_grad)
print(f'Trainable params: {n_train:,}  (GPT-2 is frozen)')
print('Sanity loss on one batch:', loss.item())

#showing 31M trainable params on MappingNetwork
#Loss of 9.029 is pretty much random right now since training has not yet happened
```

```python
# generate a caption from an audio embedding
import numpy as np, random

@torch.no_grad() #turning off gradient tracking
def caption_for(emb_vec, max_len=30):
    captioner.eval()
    v = np.asarray(emb_vec, dtype='float32') #raw CLAp embedding converted into float32 and normalizes it
    v = v / (np.linalg.norm(v) + 1e-9) #preventing divide by zero crap
    emb = torch.tensor(v, device=device).unsqueeze(0) #adding batch dimension (512,) -> (1,512)
    #building the caption one word at a time
    inputs, past, out_ids = captioner.map(emb), None, []
    for _ in range(max_len):
        o = captioner.gpt(inputs_embeds=inputs, past_key_values=past, use_cache=True)
        # new step only processes the new token and using "past" for the context for efficiency
        past = o.past_key_values
        #.argmax(-1) picking the most likely next token
        nxt = o.logits[:, -1, :].argmax(-1) #GPT2's raw scores for every possible next token; only taking the last position's score
        if nxt.item() == tok.eos_token_id: #predicting end-of-sequence token
            break
        out_ids.append(nxt.item()) #collecting the token id
        inputs = captioner.gpt.transformer.wte(nxt).unsqueeze(1) #converts back into the 768 dimension embed
    # token ids into human-readable string and removes whitespaces
    return tok.decode(out_ids).strip()

#Quick review of above with 5 random clips
for n in random.sample(all_names, 5):
    print('CLIP   :', n)
    print('  MODEL:', caption_for(X_all[name_to_row[n]]))
    print('  TRUE :', cap.loc[n, cap_cols[0]], '\n')
```

```python
# Hint builder

#Creating hints for every clip with accoustically similar 3 clips
K = 3
hint_for = {}
Xn_all = X_all.astype('float32').copy(); faiss.normalize_L2(Xn_all)
Dn, In = full_index.search(Xn_all, K + 1)          # retrieve neighbors for all clips at once
for i, name in enumerate(all_names):
    neigh = [all_names[j] for j in In[i] if all_names[j] != name][:K] #Filters out any self-matches
    hint_for[name] = 'Similar sounds: ' + ' '.join(cap.loc[nm, cap_cols[0]] for nm in neigh) + ' Caption:'
print('Example hint:\n', hint_for[all_names[0]])
```

```python
#RagDataset

#bundling each training with three things:
  #1) audio embedding
  #2) true caption
  #3) hint string
HINT_LEN, CAP_LEN = 64, 30
class RagDataset(Dataset):
    def __init__(self, df): self.df = df.reset_index(drop=True)
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        r = self.df.iloc[idx]
        emb = torch.tensor(X_all[name_to_row[r['audio_filename']]], dtype=torch.float32)
        h = tok(hint_for[r['audio_filename']], truncation=True, max_length=HINT_LEN,
                padding='max_length', return_tensors='pt')
        c = tok(r['caption'] + tok.eos_token, truncation=True, max_length=CAP_LEN,
                padding='max_length', return_tensors='pt')
        return emb, h.input_ids[0], h.attention_mask[0], c.input_ids[0], c.attention_mask[0]
rag_dl = DataLoader(RagDataset(training_df), batch_size=64, shuffle=True)
```

```python
#RagCaptioner
from torch.optim import AdamW
from tqdm import tqdm  # adding these imports here so that we can skip the cell above that we don't need to keep rerunning

# New translator being trained to use the hints
class RagCaptioner(nn.Module):
    def __init__(self):
        super().__init__()
        self.gpt = captioner.gpt           # same frozen GPT-2 as System A
        self.map = MappingNetwork()        # a NEW translator, trained WITH hints
    def forward(self, emb, hids, hmask, cids, cmask):
        prefix = self.map(emb)
        he = self.gpt.transformer.wte(hids); ce = self.gpt.transformer.wte(cids)
        inp = torch.cat([prefix, he, ce], dim=1)
        B, P = prefix.shape[0], prefix.shape[1]
        pmask = torch.ones(B, P, device=emb.device, dtype=hmask.dtype)
        mask = torch.cat([pmask, hmask, cmask], dim=1)
        labels = cids.clone(); labels[cmask == 0] = -100
        ignore = torch.full((B, P + hids.shape[1]), -100, device=emb.device, dtype=labels.dtype)
        labels = torch.cat([ignore, labels], dim=1)            # score ONLY the caption
        return self.gpt(inputs_embeds=inp, attention_mask=mask, labels=labels).loss
```

```python
@torch.no_grad()
def rag_caption(name, max_len=30):
    rag_model.eval()
    v = X_all[name_to_row[name]].astype('float32'); v = v / (np.linalg.norm(v) + 1e-9)
    emb = torch.tensor(v, device=device).unsqueeze(0)
    hids = tok(hint_for[name], truncation=True, max_length=HINT_LEN, return_tensors='pt').input_ids.to(device)
    cur = torch.cat([rag_model.map(emb), rag_model.gpt.transformer.wte(hids)], dim=1)
    past, out = None, []
    for _ in range(max_len):
        o = rag_model.gpt(inputs_embeds=cur, past_key_values=past, use_cache=True)
        past = o.past_key_values
        nxt = o.logits[:, -1, :].argmax(-1)
        if nxt.item() == tok.eos_token_id: break
        out.append(nxt.item())
        cur = rag_model.gpt.transformer.wte(nxt).unsqueeze(1)
    return tok.decode(out).strip()

for n in random.sample(all_names, 5):
    print('CLIP :', n)
    print('  A   :', caption_for(X_all[name_to_row[n]]))
    print('  B   :', rag_caption(n))
    print('  TRUE:', cap.loc[n, cap_cols[0]], '\n')
```

```python
# ===== STAGE 4: automatic metrics =====
!pip -q install pycocoevalcap pycocotools

import json, random
from pycocoevalcap.bleu.bleu import Bleu
from pycocoevalcap.meteor.meteor import Meteor
from pycocoevalcap.cider.cider import Cider

def evaluate_system(name, caption_fn, sample_names, cap, cap_cols):
    """Score a captioning system against all 5 ground truth captions."""
    refs, hyps = {}, {}
    print(f'Generating captions for {name}...')
    for i, n in enumerate(sample_names):
        hyps[i] = [caption_fn(n)]
        refs[i]  = [cap.loc[n, col] for col in cap_cols
                    if isinstance(cap.loc[n, col], str) and cap.loc[n, col].strip()]
    scorers = [
        (Bleu(4),   ['BLEU-1','BLEU-2','BLEU-3','BLEU-4']),
        (Meteor(),  ['METEOR']),
        (Cider(),   ['CIDEr']),
    ]
    scores = {}
    for scorer, names in scorers:
        score, _ = scorer.compute_score(refs, hyps)
        if isinstance(score, list):
            for n2, s in zip(names, score):
                scores[n2] = round(s, 4)
        else:
            scores[names[0]] = round(score, 4)
    return scores

# Use a fixed sample of 200 clips for eval
random.seed(42)
eval_names = random.sample(all_names, 200)

# Retrieval baseline: nearest neighbor caption (System 0)
def retrieval_caption(name):
    v = X_all[name_to_row[name]].astype('float32').copy()
    v = v / (np.linalg.norm(v) + 1e-9)
    D, I = full_index.search(v.reshape(1,-1), 2)
    nn = all_names[I[0][1]]
    return cap.loc[nn, cap_cols[0]]

# System A: direct captioning
def system_a_caption(name):
    return caption_for(X_all[name_to_row[name]])

# System B: retrieval-assisted
def system_b_caption(name):
    return rag_caption(name)

results = {}
results['Retrieval baseline'] = evaluate_system('Retrieval baseline', retrieval_caption, eval_names, cap, cap_cols)
results['System A (direct)']  = evaluate_system('System A (direct)',  system_a_caption,  eval_names, cap, cap_cols)
results['System B (RAG)']     = evaluate_system('System B (RAG)',     system_b_caption,  eval_names, cap, cap_cols)

print('\n=== Results ===')
metrics = ['BLEU-1','BLEU-2','BLEU-3','BLEU-4','METEOR','CIDEr']
header = f"{'System':<25}" + ''.join(f'{m:>10}' for m in metrics)
print(header)
print('-' * len(header))
for sys_name, scores in results.items():
    row = f"{sys_name:<25}" + ''.join(f"{scores.get(m, 0):>10.4f}" for m in metrics)
    print(row)

json.dump(results, open(os.path.join(CACHE_DIR, 'eval_results.json'), 'w'), indent=2)
print('\nSaved to cache.')
```

```python
# ===== STAGE 5: error analysis =====
import pandas as pd

random.seed(7)
analysis_names = random.sample(all_names, 50)

rows = []
for n in analysis_names:
    true_cap   = cap.loc[n, cap_cols[0]]
    pred_a     = system_a_caption(n)
    pred_b     = system_b_caption(n)
    pred_ret   = retrieval_caption(n)

    # simple word overlap as a proxy for per-clip quality
    def overlap(pred, ref):
        p, r = set(pred.lower().split()), set(ref.lower().split())
        return len(p & r) / max(len(r), 1)

    rows.append({
        'filename'   : n,
        'true'       : true_cap,
        'retrieval'  : pred_ret,
        'system_a'   : pred_a,
        'system_b'   : pred_b,
        'overlap_ret': round(overlap(pred_ret, true_cap), 3),
        'overlap_a'  : round(overlap(pred_a,   true_cap), 3),
        'overlap_b'  : round(overlap(pred_b,   true_cap), 3),
    })

df_err = pd.DataFrame(rows)

# Worst cases for System A — clips where it struggled most
print('=== 5 worst clips for System A ===')
worst_a = df_err.nsmallest(5, 'overlap_a')[['filename','true','system_a','overlap_a']]
for _, r in worst_a.iterrows():
    print(f'\nFILE  : {r.filename}')
    print(f'TRUE  : {r.true}')
    print(f'PRED  : {r.system_a}')
    print(f'SCORE : {r.overlap_a}')

# Where System B beats System A most
df_err['b_gain'] = df_err['overlap_b'] - df_err['overlap_a']
print('\n=== 5 clips where RAG helped most ===')
best_rag = df_err.nlargest(5, 'b_gain')[['filename','true','system_a','system_b','b_gain']]
for _, r in best_rag.iterrows():
    print(f'\nFILE  : {r.filename}')
    print(f'TRUE  : {r.true}')
    print(f'A     : {r.system_a}')
    print(f'B     : {r.system_b}')
    print(f'GAIN  : {r.b_gain:+.3f}')

# Where retrieval beats both models
df_err['ret_best'] = (df_err['overlap_ret'] > df_err['overlap_a']) & \
                     (df_err['overlap_ret'] > df_err['overlap_b'])
print(f'\nClips where retrieval baseline beats both models: {df_err.ret_best.sum()}/50')

df_err.to_csv(os.path.join(CACHE_DIR, 'error_analysis.csv'), index=False)
print('Saved error_analysis.csv')
```
